Данные разбиты на два листа - общая информация о точках клиента и данные о выручках клиента с 2015 года.
Клиент прислал эти данные чтобы вы, как аналитик, могли сформировать модель машинного обучения для прогнозирования выручек клиента по новым точкам. Первый этап этой работы - проверка присланных клиентом данных. 
Точки клиента разбиты на тестовую и обучающую выборки (указано на первом листе файла): по магазинам обучающей выборки клиент предоставляет свои выручки (второй лист) для обучения будущей модели. На тестовых точках клиент потом самостоятельно проверит качество прогнозирования выручек - выручки по тестовым точкам он нам не предоставляет.

Представьте, что вам дальше необходимо будет загрузить эти данные в систему и использовать для формирования модели спроса - тяжело это делать с некорректными или непонятными данными.

1. Сколько магазинов формата стрит в тестовой выборке?

```python
data['Формат магазина'][(data['Формат магазина'] == 'Street') | (data['Формат магазина'] == 'Стрит') | 
                        (data['Формат магазина'] == 'Strееt') | (data['Формат магазина'] == 'Стрт')] = 'Street'

len(data[(data['Формат магазина'] == 'Street') & (data['Выборка'] == 'Тестовая')].index)
```

2. Какова средняя выручка магазинов формата мини ТЦ за 2016 год?

```python
data['Формат магазина'][((data['Формат магазина'] == 'Мини ТЦ') | (data['Формат магазина'] == 'Мини-ТЦ'))] = 'Мини-тц'
data['Формат магазина'].value_counts()  # Единые метки для всех типов
```

```python
shopes_mini_id = data['id точки'][(data['Формат магазина'] == 'Мини-тц')  & (data['Выборка'] == 'Обучающая')].values
len(shopes_mini_id)  # Количество интересующих нас магазинов для подсчета средней выручки
```

```python
proceeds['id точки'].isna().sum()  # Для 3 магазинов с указанной выручкой отсутсвуют id
```

```python
proceeds = proceeds.dropna(subset=['id точки'])  # Удалим такие строки
proceeds['id точки'] = proceeds['id точки'].astype('int64')
proceeds = proceeds.set_index('id точки')  # Установим id точки как индексы

proceeds.head(1)
```

```python
proceeds.loc[test_index, :].shape[0]  # Количество магазинов тестовой выборки, для которых указана выручка, хотя не д.б. указана
```

```python
proceeds.dtypes  # Типы колонок датасета выручек, заметим, что нет единого формата
```

```python
columns_obj = proceeds.columns[proceeds.dtypes != 'float64']  # Для колонок, имеющих тип object переведем их в единый тип float

for column in proceeds[columns_obj]:  # Предварительный анализ показал что такие данные строковые с запятой   
    proceeds[column] = proceeds[column].replace(" ","", regex=True).replace(",",".", regex=True).astype(float)

# Поэтому убираем пробелы и заменяем ',' на '.' в таких значениях для корректного считывания
    
proceeds.dtypes
```

```python
proceeds_2 = proceeds.loc[shopes_mini_id, :]  # Датасет для интересующих нас точек

columns2016 = proceeds_2.columns[proceeds_2.columns.str.contains('2016')]  # Колонки для 2016 года

proceeds_2[columns2016]  # Видим что не для всех точек предоставлены данные по выручке
```

```python
proceeds_2[columns2016].mean().mean()  
# Считается среднее сначала по месяцам, затем среднее из подсчитанных средних - среднее за год
```
